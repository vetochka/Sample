```python
#LogisticRegression(
#       C=1.0, class_weight=None,dual=False, fit_intercept=True,intercept_scaling=1,max_iter=100,multi_class='ovr',
#       n_jobs=1, penalty='12', random_state=None, solver='liblinear', tol=0.0001, verbose=0, warm_start=False)
#fit_intercept: Булеве значення, яке визначає, чи потрібно обчислювати перехоплення (intercept)
#n_jobs: Кількість потоків, які будуть використані під час обчислення. 
#C: Зворотний коефіцієнт регуляризації. Контролює силу регуляризації: менше значення C веде до сильнішої регуляризації.
#penalty: Визначає тип регуляризації. Може бути 'l1', 'l2', 'elasticnet' або 'none'.
# 'l1' використовує регуляризацію L1 (Lasso), 'l2' - регуляризацію L2 (Ridge),
# 'elasticnet' - комбінацію L1 та L2, 'none' - без регуляризації.
#solver: Визначає алгоритм оптимізації. Доступні варіанти включають 'newton-cg', 'lbfgs', 'liblinear', 'sag' і 'saga'.
# Вибір алгоритму залежить від розміру даних та типу регуляризації.
#random_state: Параметр, який визначає випадкову початкову точку для алгоритму.
# Використання фіксованого значення дозволяє забезпечити відтворюваність результатів.
```

```python
result = {'model_name':[],'model_config':[],'train_score':[],'test_score':[],'accuracy':[]}
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
for penalty in ('l1','l2','elasticnet','none'):
    for solver in ('newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'):
        for c in (0.1, 1.0, 10.0, 100.0, 1000.0):
            if ((solver in ('newton-cg', 'lbfgs', 'sag') and penalty in ('l2', 'none')) or
                (solver == 'liblinear' and penalty in ('l1', 'l2')) or #AI fix penalty = 'none' -> error 
                (solver == 'saga' and penalty == 'elasticnet')):
                model = LogisticRegression(penalty=penalty, C=c, solver=solver, l1_ratio=0.5 if penalty == 'elasticnet' else None) 
                model.fit(X_train, y_train)
                # Оцінка моделі
                train_score = model.score(X_train, y_train)
                test_score = model.score(X_test, y_test)
                # Розрахунок помилки
                y_predict = model.predict(X_test)
                #викликає помилку rmse = mean_squared_error(y_test, y_predict)**0.5
                # accuracy - % правильно "вгаданих" значень 
                accuracy = accuracy_score(y_test, y_predict)
                # Збереження результатів
                result['model_name'].append('LogisticRegression')
                result['model_config'].append(f"Penalty={penalty}, solver={solver}, C={c}")
                result['train_score'].append(train_score)
                result['test_score'].append(test_score)
                result['accuracy'].append(accuracy)
pd.DataFrame(result)
```
